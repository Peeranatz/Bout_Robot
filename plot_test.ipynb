```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#  CSV
df = pd.read_csv("gimbal_tof_angle_data_2.csv")

# 
df['distance_cm'] = df['distance'] / 10.0

#  Yaw angle 
df['yaw_rad'] = np.deg2rad(df['yaw_angle'])

#  x, y 
df['y'] = df['distance_cm'] * np.cos(df['yaw_rad'])
df['x'] = df['distance_cm'] * np.sin(df['yaw_rad'])

#  ()
plt.figure(figsize=(10, 6))
plt.scatter(df['x'], df['y'], marker='o',color='red', alpha=0.7)
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.title('Vector Points from Yaw and Distance (cm)')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#  CSV
df = pd.read_csv("gimbal_tof_angle_data_2.csv")

# 
df['distance_cm'] = df['distance'] / 10.0

#  Yaw angle 
df['yaw_rad'] = np.deg2rad(df['yaw_angle'])

#  x, y 
df['y'] = df['distance_cm'] * np.cos(df['yaw_rad'])
df['x'] = df['distance_cm'] * np.sin(df['yaw_rad'])

# 
plt.figure(figsize=(14, 10))

#  (x, y)
plt.quiver(
    np.zeros(len(df)),          #  x = 0
    np.zeros(len(df)),          #  y = 0
    df['x'],                    #  x
    df['y'],                    #  y
    angles='xy', scale_units='xy', scale=1,
    color='blue', alpha=0.6
)

# 
plt.scatter(df['x'], df['y'], color='red', s=40, label='End Points')

# 
plt.xlabel('X (cm)', fontsize=14)
plt.ylabel('Y (cm)', fontsize=14)
plt.title('Vectors from Center to Each Point (with End Points)', fontsize=16)
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()

```
