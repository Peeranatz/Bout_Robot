```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# โหลดข้อมูลจาก CSV
df = pd.read_csv("gimbal_tof_angle_data_2.csv")

# แปลงระยะทางจากมิลลิเมตรเป็นเซนติเมตร
df['distance_cm'] = df['distance'] / 10.0

# แปลง Yaw angle จากองศาเป็นเรเดียน
df['yaw_rad'] = np.deg2rad(df['yaw_angle'])

# คำนวณพิกัด x, y จากเวกเตอร์
df['y'] = df['distance_cm'] * np.cos(df['yaw_rad'])
df['x'] = df['distance_cm'] * np.sin(df['yaw_rad'])

# วาดกราฟ
plt.figure(figsize=(14, 10))

# วาดเวกเตอร์จากจุดศูนย์กลางไปยัง (x, y)
plt.quiver(
    np.zeros(len(df)),          # จุดเริ่มต้น x = 0
    np.zeros(len(df)),          # จุดเริ่มต้น y = 0
    df['x'],                    # ทิศทางแนว x
    df['y'],                    # ทิศทางแนว y
    angles='xy', scale_units='xy', scale=1,
    color='blue', alpha=0.6
)

# วาดจุดที่ปลายแต่ละเวกเตอร์
plt.scatter(df['x'], df['y'], color='red', s=40, label='End Points')

# ตั้งค่ากราฟ
plt.xlabel('X (cm)', fontsize=14)
plt.ylabel('Y (cm)', fontsize=14)
plt.title('Vectors from Center to Each Point (with End Points)', fontsize=16)
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()

```
