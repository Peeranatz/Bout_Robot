Robotmaster Attitude Data 

Pitch,Roll and Yaw Value

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv(r"csv\robomaster_attitude_data.csv")

# Create figure
plt.figure(figsize=(10, 6))

# Plot pitch, roll, yaw over time
plt.plot(data["timestamp"], data["pitch_deg"], label="Pitch ()")
plt.plot(data["timestamp"], data["roll_deg"], label="Roll ()")
plt.plot(data["timestamp"], data["yaw_deg"], label="Yaw ()")

# Add labels and title
plt.xlabel("Time (s)")
plt.ylabel("Angle (degrees)")
plt.title("Pitch, Roll, Yaw over Time")
plt.legend()
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```

Robomaster ESC Data



```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv(r"csv\robomaster_esc_data.csv")

fig, axs = plt.subplots(4, 1, figsize=(12, 10), sharex=True)

# Plot esc_state_fr
axs[0].plot(data["timestamp"], data["esc_state_fr"], color="tab:blue")
axs[0].set_ylabel("ESC State")
axs[0].set_title("ESC State Front-Right")

# Plot esc_speed_fr
axs[1].plot(data["timestamp"], data["esc_speed_fr"], color="tab:orange")
axs[1].set_ylabel("Speed")
axs[1].set_title("ESC Speed Front-Right")

# Plot esc_angle_fr
axs[2].plot(data["timestamp"], data["esc_angle_fr"], color="tab:green")
axs[2].set_ylabel("Angle")
axs[2].set_title("ESC Angle Front-Right")

# Plot esc_current_fr
axs[3].plot(data["timestamp"], data["esc_current_fr"], color="tab:red")
axs[3].set_ylabel("Current")
axs[3].set_title("ESC Current Front-Right")
axs[3].set_xlabel("Time (s)")

plt.tight_layout()
plt.show()
```

Robomaster IMU Data 

```python
import pandas as pd
import matplotlib.pyplot as plt

# 
data = pd.read_csv(r"csv\robomaster_imu_data.csv")

plt.figure(figsize=(12, 6))

#  Acc XYZ
plt.plot(data["timestamp"], data["acc_x_g"], label="Acc X (g)")
plt.plot(data["timestamp"], data["acc_y_g"], label="Acc Y (g)")
plt.plot(data["timestamp"], data["acc_z_g"], label="Acc Z (g)")

plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Accelerometer Data over Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Robomaster Position Data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv(r"csv\robomaster_position_data.csv")

# Create figure
plt.figure(figsize=(10, 6))

# Plot X, Y, Z position over time
plt.plot(data["timestamp"], data["pos_x_m"], label="X Position (m)")
plt.plot(data["timestamp"], data["pos_y_m"], label="Y Position (m)")
plt.plot(data["timestamp"], data["pos_z_deg_from_start"], label="Z Position (m)")

# Add labels and title
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("X, Y, Z Position over Time")
plt.legend()
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```

Robomaster Status Data

```python
import pandas as pd
import matplotlib.pyplot as plt

# 
data = pd.read_csv(r"csv\robomaster_status_data.csv")  # 

plt.figure(figsize=(12, 4))
plt.step(
    data["timestamp"], data["status_chassis_is_moving"], where="post", label="Is Moving"
)

plt.xlabel("Time (s)")
plt.ylabel("Moving Status (1 = stop, 0 = moving)")
plt.title("Chassis Moving Status Over Time")
plt.ylim(-0.1, 1.1)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
