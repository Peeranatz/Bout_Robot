#### TEAM BOUNDING BOUT


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("chicken_tracking.csv")

#  mapping  -> 
label_map = {
    'Error_Total': 'err_total(t)',
    'Error_Yaw': 'err_x(t)',
    'Error_Pitch': 'err_y(t)',
    'Accumulate_Error_X': 'accumulate_err_x(t)',
    'Accumulate_Error_Y': 'accumulate_err_y(t)',
    'Accumulate_Error_Total': 'accumulate_err_total(t)',
    'Controller_Output_X': 'controller_output_x(t)',
    'Controller_Output_Y': 'controller_output_y(t)',
    'Gimbal_Pitch_Angle': 'gimbal_pitch_angle(t)',
    'Gimbal_Yaw_Angle': 'gimbal_yaw_angle(t)'
}

for col, label in label_map.items():
    plt.figure(figsize=(10,4))
    plt.plot(df['Timestamp'], df[col], label=label, color='b')
    plt.title(f'{label} vs Time')
    plt.xlabel('Timestamp (s)')
    plt.ylabel(label)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
