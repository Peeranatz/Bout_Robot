Robotmaster Attitude Data Sensor

Pitch,Roll and Yaw Value

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv(r"csv_3\robomaster_attitude_data.csv")

# Create figure
plt.figure(figsize=(10, 6))

# Plot pitch, roll, yaw over time
plt.plot(data["timestamp"], data["pitch_deg"], label="Pitch ()")
plt.plot(data["timestamp"], data["roll_deg"], label="Roll ()")
plt.plot(data["timestamp"], data["yaw_deg"], label="Yaw ()")

# Add labels and title
plt.xlabel("Time (s)")
plt.ylabel("Angle (degrees)")
plt.title("Pitch, Roll, Yaw over Time")
plt.legend()
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```

Robomaster ESC Data Sensor

State,Speed and Angle Data



```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv(r"csv_3\robomaster_esc_data.csv")

fig, axs = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

# Plot esc_state_fr
axs[0].plot(data["timestamp"], data["esc_state_fr"], color="tab:blue")
axs[0].set_ylabel("ESC State")
axs[0].set_title("ESC State Front-Right Wheel")

# Plot esc_speed_fr
axs[1].plot(data["timestamp"], data["esc_speed_fr"], color="tab:orange")
axs[1].set_ylabel("Speed")
axs[1].set_title("ESC Speed Front-Right Wheel")

# Plot esc_angle_fr
axs[2].plot(data["timestamp"], data["esc_angle_fr"], color="tab:green")
axs[2].set_ylabel("Angle")
axs[2].set_title("ESC Angle Front-Right Wheel")

plt.tight_layout()
plt.show()
```

Robomaster Status Data

```python
import pandas as pd
import matplotlib.pyplot as plt

#  ( path )
data = pd.read_csv(r"csv_3\robomaster_status_data.csv")

#  plot
status_columns = [
    "status_static_flag",
    "status_up_hill",
    "status_down_hill",
    "status_on_slope",
    "status_pick_up",
    "status_impact_x",
    "status_impact_y",
    "status_impact_z",
    "status_roll_over",
    "status_hill_static",
    "status_val_10"
]

#  dataframe  ( error)
available_columns = [col for col in status_columns if col in data.columns]

plt.figure(figsize=(15, 6))

#  plot 
for col in available_columns:
    plt.step(data["timestamp"], data[col], where="post", label=col)

plt.xlabel("Time (s)")
plt.ylabel("Status Flag (0 or 1)")
plt.title("Robot Status Flags Over Time")
plt.ylim(-0.1, 1.1)
plt.legend(loc="upper right", bbox_to_anchor=(1.15, 1.0))
plt.grid(True)
plt.tight_layout()
plt.show()
```
