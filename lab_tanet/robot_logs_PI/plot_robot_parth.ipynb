```python
import pandas as pd
import matplotlib.pyplot as plt

# 
files = [
    ("robot_path_log_round_0.5I1.csv", "Round 1"),
    ("robot_path_log_round_0.5I2.csv", "Round 2"),
    ("robot_path_log_round_0.5I3.csv", "Round 3"),
    ("robot_path_log_round_0.5I4.csv", "Round 4")
]

# 
colors = ['blue', 'green', 'red', 'orange']

# 
plt.figure(figsize=(8, 8))

# 
for (file_path, label), color in zip(files, colors):
    df = pd.read_csv(file_path)
    plt.plot(df['X (m)'], df['Y (m)'], marker='o', linestyle='-', color=color, label=label)

# 
plt.title("Robot Path With PI 0.5 Comparison per Round")
plt.xlabel("X (m)")
plt.ylabel("Y (m)")
plt.grid(True)
plt.axis('equal')  #  X/Y 
plt.legend()       #  Legend 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ========  path  CSV  ========
files = [
    "robot_path_log_round_0.5I1.csv"
]

#  plot
columns_to_plot = ["P", "I", "D", "Speed", "Error (m)"]  # 
time_column = "Time (s)"  # 

#  subplot
fig, axs = plt.subplots(len(columns_to_plot), 1, figsize=(12, 10), sharex=True)
fig.suptitle("PID, SPEED, ERROR WITH PD CONTROLLER Ki VALUE 0.5", fontsize=16)

#  (P, I, D, SPEED, ERROR)
for idx, col in enumerate(columns_to_plot):
    ax = axs[idx]
    for file in files:
        df = pd.read_csv(file)
        ax.scatter(df[time_column], df[col], label=file)
    ax.set_ylabel(col)
    ax.legend()
    ax.grid(True)

axs[-1].set_xlabel("Time")
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
