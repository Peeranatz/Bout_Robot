```python
import pandas as pd
import matplotlib.pyplot as plt

# รายชื่อไฟล์และชื่อรอบที่ต้องการแสดงผล
files = [
    ("robot_path_log_round_BEST_PID1.csv", "Round 1"),
    ("robot_path_log_round_BEST_PID2.csv", "Round 2"),
    ("robot_path_log_round_BEST_PID3.csv", "Round 3"),
    ("robot_path_log_round_BEST_PID4.csv", "Round 4")
]

# สีที่ใช้สำหรับแต่ละรอบ
colors = ['blue', 'green', 'red', 'orange']

# เริ่มพล็อต
plt.figure(figsize=(8, 8))

# วนลูปพล็อตแต่ละรอบ
for (file_path, label), color in zip(files, colors):
    df = pd.read_csv(file_path)
    plt.plot(df['X (m)'], df['Y (m)'], marker='o', linestyle='-', color=color, label=label)

# ตกแต่งกราฟ
plt.title("Robot Path With PID Best Case Comparison per Round")
plt.xlabel("X (m)")
plt.ylabel("Y (m)")
plt.grid(True)
plt.axis('equal')  # ทำให้สัดส่วนแกน X/Y เท่ากัน
plt.legend()       # แสดง Legend แยกรอบ
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ======== ใส่ path ไฟล์ CSV ของคุณ ========
files = [
    "robot_path_log_round_BEST_PID1.csv"
]

# ชื่อคอลัมน์ที่ต้องการ plot
columns_to_plot = ["P", "I", "D", "Speed", "Error (m)"]  # ปรับชื่อให้ตรงกับในไฟล์
time_column = "Time (s)"  # ปรับชื่อคอลัมน์เวลาให้ตรง

# สร้าง subplot
fig, axs = plt.subplots(len(columns_to_plot), 1, figsize=(12, 10), sharex=True)
fig.suptitle("PID, SPEED, ERROR WITH PID CONTROLLER BEST CASE", fontsize=16)

# วนแต่ละกราฟ (P, I, D, SPEED, ERROR)
for idx, col in enumerate(columns_to_plot):
    ax = axs[idx]
    for file in files:
        df = pd.read_csv(file)
        ax.scatter(df[time_column], df[col], label=file)
    ax.set_ylabel(col)
    ax.legend()
    ax.grid(True)

axs[-1].set_xlabel("Time")
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
